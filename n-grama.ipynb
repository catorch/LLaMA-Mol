```python
def construir_bigramas_de_molecula(molecula):
    lista_bigramas = []
    for i in range(len(molecula)):
        if i == len(molecula) - 1:
            break
        bigrama = (molecula[i],molecula[i + 1])
        lista_bigramas.append(bigrama)

    return lista_bigramas
```

```python
# Iterar moleculas
bigramas_calculados = []
for molecula in corpus_tokenizado:
    bigramas_molecula = construir_bigramas_de_molecula(molecula)
    bigramas_calculados.append(bigramas_molecula)

bigramas_calculados
```

```python
bigramas_aplanados = []

for bigramas_molecula in bigramas_calculados:
    for bigrama in bigramas_molecula:
        bigramas_aplanados.append(bigrama)

bigramas_aplanados
```

```python
# Contar numero de repeticiones de bigramas

conteo_bigramas = {}

for bigrama in bigramas_aplanados:
    
    if bigrama in conteo_bigramas:
        conteo_bigramas[bigrama] += 1
    else:
        conteo_bigramas[bigrama] = 1

conteo_bigramas

```

```python
# Contar numero de repeticiones unigrama
conteo_unigramas = {}

for molecula in corpus_tokenizado:
    for atomo in molecula:
        
        if atomo in conteo_unigramas:
            conteo_unigramas[atomo] += 1
        else:
            conteo_unigramas[atomo] = 1

conteo_unigramas
```

```python
# Calcular probabilidades de los bigramas
probabilidades_bigramas = {}
for bigrama in conteo_bigramas.keys():
    atomo1, atomo2 = bigrama
    probabilidades_bigramas[bigrama] = conteo_bigramas[bigrama] / conteo_unigramas[atomo1]

probabilidades_bigramas
    
```

```python
import random

def generar_siguiente_atomo(atomo_actual, probabilidades_bigramas):
          
    bigramas_candidatos = []

    # Filtrar para obtener posibles bigramas
    for bigrama in probabilidades_bigramas.keys():
        if bigrama[0] == atomo_actual:
            bigramas_candidatos.append(bigrama)

    # Verificar si hay bigramas posibles (>0)
    if len(bigramas_candidatos) == 0:
        return ""

    # Obtener probabilidades de los bigramas_candidatos
    probabilidades = []
    for bigrama in bigramas_candidatos:
        probabilidades.append(probabilidades_bigramas[bigrama])

    # Seleccionar bigrama aleatoriamente
    bigrama_elegido = random.choices(bigramas_candidatos, weights=probabilidades, k=1)[0]

    return bigrama_elegido[1]

generar_siguiente_atomo('C', probabilidades_bigramas)
```

```python
# Ahora, generamos una cadena SMILES a partir del modelo de bigramas
def generar_molecula(molecula, probabilidades_bigramas, max_atomos=6):
    
    while True:
        atomo_actual = molecula[-1]    
        siguiente_atomo = generar_siguiente_atomo(atomo_actual, probabilidades_bigramas)
        
        if not siguiente_atomo or siguiente_atomo == "":
            break

        molecula += siguiente_atomo

        if len(molecula) == max_atomos:
            break

    return molecula
        
```

```python
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski

def lipinski_rule_of_five(mol):
    """
    Evaluate a molecule against Lipinski's Rule of Five.
    """
    # Calculate the descriptors
    mol_weight = Descriptors.MolWt(mol)
    logP = Descriptors.MolLogP(mol)
    h_donors = Lipinski.NumHDonors(mol)
    h_acceptors = Lipinski.NumHAcceptors(mol)
    
    # Check against the rules
    criteria = [
        mol_weight <= 500,
        logP <= 5,
        h_donors <= 5,
        h_acceptors <= 10
    ]
    
    return criteria

def prepare_data(smiles_string):
    # Create a new DataFrame with a single row for the SMILES string
    data = {'canonical_smiles': [smiles_string]}
    new_df = pd.DataFrame(data)
    
    # Generate a molecule object
    mol = Chem.MolFromSmiles(smiles_string)
    if not mol:
        raise ValueError("Invalid SMILES string provided.")
    
    # Apply Lipinski's Rule of Five and unpack results into separate columns
    new_df['MW_check'], new_df['LogP_check'], new_df['HBD_check'], new_df['HBA_check'] = zip(*[lipinski_rule_of_five(mol)])
    
    # Generate ECFP4 fingerprint
    ecfp4 = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=2048)
    ecfp4_bits = list(ecfp4.ToBitString())
    
    new_df['ECFP4'] = [ecfp4_bits]
    
    # Convert the 'ECFP4' column to a DataFrame
    ecfp4_df = new_df['ECFP4'].apply(pd.Series)
    
    # Rename the columns
    ecfp4_df.columns = [f'bit_{i}' for i in range(2048)]
    
    # Concatenate the original DataFrame with the new ECFP4 columns
    expanded_df = pd.concat([new_df, ecfp4_df], axis=1)
    
    # Drop the 'canonical_smiles' and 'ECFP4' columns
    expanded_df.drop(columns=['canonical_smiles', 'ECFP4'], inplace=True)
    # print(expanded_df)
    # Return the processed DataFrame row
    return expanded_df.copy()

# Example
smiles_string = "CCO"
prepared_data = prepare_data(smiles_string)
prepared_data.head()
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
import py3Dmol

def show_molecule_3d(mol):
    mb = Chem.MolToMolBlock(mol)
    viewer = py3Dmol.view(width=400, height=400)
    viewer.addModel(mb, format="mol")
    viewer.setStyle({'stick': {}})
    viewer.zoomTo()
    return viewer.show()
    
# Add Hydrogens to the molecule
mol = Chem.AddHs(mol)

# Generate 3D coordinates
AllChem.EmbedMolecule(mol, AllChem.ETKDG())

# Optimize the molecule using a force field (e.g., MMFF94)
AllChem.MMFFOptimizeMolecule(mol, maxIters=500, nonBondedThresh=500.0)

# Visualize
show_molecule_3d(mol)
```

```python
import requests

def search_pubchem(smiles, property):
    base_url = "https://pubchem.ncbi.nlm.nih.gov/rest/pug"
    url = f"{base_url}/compound/smiles/{smiles}/property/{property}/JSON"
    response = requests.get(url)
    data = response.json()
    return data


property = "MolecularWeight"
result = search_pubchem(nueva_molecula, property)
print(f"The {property} of molecule {nueva_molecula} is: {result}")
```
