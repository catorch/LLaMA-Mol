```python
import re

# Lists to store extracted iteration numbers and loss values
iterations = []
losses = []

# Regular expression pattern to extract iteration number and loss value
pattern = re.compile(r'iter (\d+): loss ([\d.]+)')

# Read the file and extract the required values
with open("training_output.txt", "r") as file:
    for line in file:
        match = pattern.search(line)
        if match:
            iterations.append(int(match.group(1)))
            losses.append(float(match.group(2)))

iterations[:10], losses[:10]  # Display first 10 values of each list
```

```python
import matplotlib.pyplot as plt

# Plotting the loss values
plt.figure(figsize=(12, 6))
plt.plot(iterations, losses, label="Training Loss", color="blue")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.title("Training Loss per Iteration")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Adjust the y-axis limits to center the scale between 0.5 and 1
plt.figure(figsize=(12, 6))
plt.plot(iterations, losses, label="Training Loss", color="blue")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.title("Training Loss per Iteration")
plt.ylim(0.5, 1)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Lists to store extracted iteration numbers and validation losses
val_iterations = []
val_losses = []

# Regular expression pattern to extract iteration number and validation loss value
val_pattern = re.compile(r'step (\d+): val loss ([\d.]+)')

# Read the file and extract the required values
with open("training_output.txt", "r") as file:
    for line in file:
        match = val_pattern.search(line)
        if match:
            val_iterations.append(int(match.group(1)))
            val_losses.append(float(match.group(2)))

val_iterations, val_losses  # Display the extracted values
```

```python
# Plotting the validation loss values
plt.figure(figsize=(12, 6))
plt.plot(val_iterations, val_losses, label="Validation Loss", color="green", marker='o')
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.title("Validation Loss per Iteration")
plt.ylim(0.5, 0.7)  # Keeping the y-axis scale same as before
plt.legend()
plt.grid(True)
plt.show()
```
